#### Problema 2
2. $y - 5 = 12$
* Agrupar: Aquí no hay términos semejantes del mismo lado.
* Resolver: Sumamos $5$ a ambos lados para despejar la $y$.$y = 12 + 5$
* Resultado: $y = 17$

```python
from sympy import symbols, Eq, solve

y = symbols('y')
# Ecuación: y - 5 = 12
ecuacion = Eq(y - 5, 12)
solucion = solve(ecuacion, y)
print(f"Resultado y: {solucion[0]}")
```

#### Problema  4
 $12 = m + 3m$
 * Agrupar: Sumamos los términos con $m$ en el lado derecho ($1m + 3m = 4m$).
 * Ecuación: $12 = 4m$
 * Resolver: Dividimos ambos lados entre $4$.
 * Resultado: $m = 3$

```python
from sympy import symbols, Eq, solve

m = symbols('m')
# Ecuación: 12 = m + 3m
ecuacion = Eq(12, m + 3*m)
solucion = solve(ecuacion, m)
print(f"Resultado m: {solucion[0]}")
```

#### Problema 6
$5b + 2b - 3b = 16$
* Agrupar: Sumamos y restamos los términos con $b$ ($5 + 2 - 3 = 4$).
* Ecuación: $4b = 16$
* Resolver: Dividimos ambos lados entre $4$.
* Resultado: $b = 4$

```python
from sympy import symbols, Eq, solve

b = symbols('b')
# Ecuación: 5b + 2b - 3b = 16
ecuacion = Eq(5*b + 2*b - 3*b, 16)
solucion = solve(ecuacion, b)
print(f"Resultado b: {solucion[0]}")
```

#### Problema 8
$7y - 5 = 3y + 7$
* Agrupar: Movemos las $y$ a un lado y los números al otro. Restamos $3y$ de ambos lados y sumamos $5$ a ambos lados.$7y - 3y = 7 + 5$$4y = 12$
* Resolver: Dividimos entre $4$.
* Resultado: $y = 3$

```python
from sympy import symbols, Eq, solve

y = symbols('y')
# Ecuacion: 7y - 5 = 3y + 7
ecuacion = Eq (7*y - 5, 3*y + 7)
solucion = solve(ecuacion, y)
print(f"Resultado y: {solucion[0]}")
```

#### Problema 10
$8m - 6 = 4n + 10$.
* Agrupar: Restamos $4m$ de ambos lados y sumamos $6$.$8m - 4n = 10 + 6$$4m = 16$
* Resolver: Dividimos entre $4$.
* Resultado: $n = 4$

```python
from sympy import symbols, Eq, solve

m = symbols('m')
# Ecuación: 8m - 6 = 4m + 10
ecuacion = Eq(8*m - 6, 4*m + 10)
solucion = solve(ecuacion, m)
print(f"Resultado n: {solucion[0]}")
```

#### Problema 12
$$10b - 4b - 8 = 10$$Esto se simplifica a:$$6b - 8 = 10$$
* Agrupar los términos constantes (números):  $-8$ al lado derecho. Para hacerlo, sumamos $8$ a ambos lados:$$6b = 10 + 8$$Esto se simplifica a:$$6b = 18$$Despejar la variable:Finalmente, como el $6$ está multiplicando a $b$, dividimos ambos lados entre $6$:$$b = \frac{18}{6}$$Resultado:$$b = 3$$

```python
from sympy import symbols, Eq, solve

# Definimos la variable simbólica
b = symbols('b')

# Definimos la ecuación: 10b - 8 = 4b + 10
# Eq(lado_izquierdo, lado_derecho)
ecuacion = Eq(10*b - 8, 4*b + 10)

# Resolvemos la ecuación para b
solucion = solve(ecuacion, b)

# Mostramos el resultado
print(f"El valor de b es: {solucion[0]}")
```
